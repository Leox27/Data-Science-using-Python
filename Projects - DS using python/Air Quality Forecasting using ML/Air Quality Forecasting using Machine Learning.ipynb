Loading the dataset of the Air Quality Forecasing that is derived from the website uc irvine.

Dataset Link: https://archive.ics.uci.edu/dataset/360/air+quality

```python
air_quality = pd.read_csv("/content/AirQualityUCI.csv", sep=";", decimal=",")
air_quality
```

We seen in this, downside of the dataset some rows are totally null.

Let fix this with by removing it because these rows are totally empty.

```python
air_quality.iloc[[9357]]
```

```python
air_quality.head(9356)
```

```python
air_quality.info()
```

Now we have to remove those "-200" values that had been added this data set randomly as the publisher describes in the information phase.

```python
air_quality.isnull().sum()
```

```python
air_quality = air_quality.replace(to_replace=-200, value=np.nan)
air_quality
```

The dataset describes that "Missing values are tagged with -200 value." So we have to replace that "-200" values with "mean".

```python
air_quality.isnull().sum()
```

```python
air_quality.drop(columns=["Date", "Time"]).mean()
```

```python
air_quality = air_quality.fillna(air_quality.drop(columns=["Date", "Time"]).mean())
air_quality
```

```python
air_quality.isnull().sum()
```

```python
air_quality.shape
```

```python
air_quality.info()
```

# Below we are now traing model based on this dataset #
There are two algorithm we can use are:

1. Forecasting (using FB Prophet model)
2. Regression

As we choose the Forecasting (FB Prophet model) for that we have to ideally YYYY-MM-DD for a date or YYYY-MM-DD HH:MM:SS for a timestamp.

Model link: https://facebook.github.io/prophet/docs/quick_start.html

```python
date_info = pd.to_datetime(
    air_quality['Date'],
    format='%d/%m/%Y'
)
```

```python
date_info
```

```python
time_info = air_quality['Time']
time_info
```

```python
time_info = time_info.apply(lambda x:x.replace('.', ':'))
```

```python
time_info
```

Let's see the pandas date_info and time_info

```python
print(type(date_info))
print(type(time_info))
```

#combining these thwo columns date_info and time_info in single one column

```python
date_time = pd.concat([date_info, time_info], axis=1)
date_time
```

```python
date_time.shape
```

```python
date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
date_time
```

```python
date_time.info()
```

Now we have to convert the ds's 'object' Dtype to 'datetime64[ns]' Dtype

```python
data = pd.DataFrame()
```

```python
data['ds']  =  pd.to_datetime(date_time['ds'])
data
```

```python
data.info()
```

Now the 2nd column for the Phophet is 'y'

```python
data['y'] = air_quality['CO(GT)']
data
```
